```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
path = f'imgs/000253.png'
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
```

```python
thickness = 2

# gpt4v result
satrt_point_gpt = (300, 50)
ending_point_gpt = (450, 275)
color_gpt = (0, 0, 255)
box_gpt = cv2.rectangle(img, satrt_point_gpt, ending_point_gpt, color_gpt, thickness)

# ground truth
satrt_point = (285, 8)
ending_point = (500, 333)
color = (255, 0, 0)
box_truth = cv2.rectangle(img, satrt_point, ending_point, color, thickness)

plt.imshow(box_truth)
```
